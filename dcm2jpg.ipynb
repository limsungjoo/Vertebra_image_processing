```python
import numpy as np
from PIL import Image
from glob import glob
import SimpleITK as sitk
```

```python
def dcm2jpg(dcm,img_path):
	image = sitk.ReadImage(dcm)
	dicom = sitk.GetArrayFromImage(image)[0]
	dicom = (dicom - dicom.min()) / (dicom.max() - dicom.min()) * 255
	dicom = np.stack((dicom,)*3,axis=-1).astype("uint8")

	im = Image.fromarray(dicom)
	im.save(img_path)

	return


file_list = glob("./img_dcm/*")

for num,dcm_file in enumerate(file_list):
# 	print("Non-VF",num+1,"/",len(file_list))
	img_path = "./img_jpg/"+dcm_file.split("/")[-1]
	print(img_path)
	img_path = img_path.replace(".dcm",".jpg")

	try:
		dcm2jpg(dcm_file,img_path)
	except:
		print(img_path)



# file_list = glob("VF/*/*")

# for num,dcm_file in enumerate(file_list):
# 	print("VF",num+1,"/",len(file_list))
	
# 	img_path = "img_jpg/VF/"+dcm_file.split("/")[-2]+"_"+dcm_file.split("/")[-1]
# 	img_path = img_path.replace(".dcm",".jpg")

# 	try:
# 		dcm2jpg(dcm_file,img_path)
# 	except:
# 		print(img_path)
```

```python
import os
import cv2
mask_path = './mask'
image_path = './img_jpg'
save_path = './img_jpg_sample'
for img in os.listdir(image_path):
    img_name = img.split('.')[-2]
    print(img_name)
    for msk in os.listdir(mask_path):
        msk_name = msk[:-10]
        print(msk_name)
        if img_name ==msk_name:
            image = cv2.imread(os.path.join(image_path,img),0)
            cv2.imwrite(os.path.join(save_path,img),image)
```
